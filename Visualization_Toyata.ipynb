1. Frequency Table:

* To check relationship b/w variables means what frequency of each category that are available under a variable.
* craete it using "pd.crosstab()" 

```python
pd.crosstab(index = df['FuelType'], columns = 'Count', dropna = True)
```

```python
df['FuelType'].isna().sum()
```

2. Two Way Table

* To check relationship b/w 2 categorical variables.
* Here we will look at the frequency distribution of gearbox types w.r.t diff. FuelType of cars.

```python
pd.crosstab(index = df['Automatic'], columns = df['FuelType'], dropna = True)
```

3. Two Way Table - Joint Probability

* it is likelihood of e independent events happening at the same time
* normalize is by default, normalize = True turn it into proportions

```python
pd.crosstab(index = df['Automatic'], columns = df['FuelType'],normalize = True, dropna = True)
```

4. Two Way Table - Marginal Probability

* It is the probability of occurance of single event
* margins is false by default, margins True gives probability of a single event regardless of the other event

```python
pd.crosstab(index = df['Automatic'], columns = df['FuelType'],margins = True, normalize = True, dropna = True)
```

4. Two Way Table - Conditional Probability

* It is probability of an event(A) given that another event(B) has already occured.
* Given gearbox what are the probability of different fueltypes.
* value in normalize is considered the first event and the the second event is on the base on it

```python
pd.crosstab(index = df['Automatic'], columns = df['FuelType'], normalize = 'index', dropna = True)
```

```python
pd.crosstab(index = df['Automatic'], columns = df['FuelType'], normalize = 'columns', dropna = True)
```
